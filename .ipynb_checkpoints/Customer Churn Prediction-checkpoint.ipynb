## Outlier Detection

```python
cols = ['AccountAge', 'MonthlyCharges', 'TotalCharges', 'ViewingHoursPerWeek',
       'AverageViewingDuration', 'ContentDownloadsPerMonth', 'UserRating',
       'SupportTicketsPerMonth', 'WatchlistSize']
```

IQR = Q3-Q1
* Lower limit = Q1-1.5*IQR
* Upprer limit = Q3+1.5*IQR

```python
Q1 = train_df[cols[1]].quantile(0.25)
Q3 = train_df[cols[1]].quantile(0.75)
Q1,Q3
```

```python
IQR = Q3-Q1
IQR
```

```python
Lower = Q1-1.5*IQR
Upper = Q3+1.5*IQR
Lower,Upper
```

```python
def detect_outiler(col):
    Limits = {}
    for i in col:
        Q1 = train_df[i].quantile(0.25)
        Q3 = train_df[i].quantile(0.75)
        IQR = Q3-Q1
        Lower = Q1-1.5*IQR
        Upper = Q3+1.5*IQR
        max_ = train_df[i].max()
        Limits[i] = {"Lower":Lower,"Upper":Upper,"MAx":max_}
    return Limits
        
```

```python
detect_outiler(cols)
```

```python
def plot_boxplots(df, cols):
    for col in cols:
        plt.figure(figsize=(6,4))
        sns.boxplot(y=df[col])
        plt.title(f'Boxplot of {col}')
        plt.show()
```

```python
plot_boxplots(train_df,cols)
```

### It is evident that dataset has no outlier

```python
def age_group(age):
    if age<=0:
        return "Not Relevent"
    if age>0 and age<=10:
        return "Children"
    if age>10 and age<=19:
        return "Teenager"
    if age >19 and age <=25:
        return "Students"
    if age >25 and age<= 32:
        return "Bechlors"
    if age >32 and age<=50:
        return "Married"
    if age >50 and age<=65:
        return "Retired"
    if age>65 and age <=85:
        return "Old"
    if age> 85:
        return "Very Old"
    
train_df["Age Group"] = train_df["AccountAge"].apply(age_group)
train_df
```

### Using Chi-Square test to determine relationship of categorical variable with churn

```python
from scipy.stats import chi2_contingency
ct = pd.crosstab(train_df["SubscriptionType"],train_df["Churn"])
chi2, p, dof, expected = chi2_contingency(ct)
formatted = f"{p:.2e}"
print(p)
```

```python
def catgory_churn_relation(cols):
    results = []
    for col in cols:
        ct = pd.crosstab(train_df[col],train_df["Churn"])
        chi2, p, dof, expected = chi2_contingency(ct)
        results.append({"Feature":col,
                     "p_value":p})
        df = pd.DataFrame(results)
    return df
    
```

```python
cols_cat = train_df.select_dtypes(include=['object']).columns.to_list()
cols_cat.remove("CustomerID")
print(cols_cat)
```

```python
df = catgory_churn_relation(cols_cat)
df["greater than 0.05"] = df["p_value"].apply(lambda x:"remove" if x>0.05 else "No")
df
```

## VIF-- To determine Multicolinearity amoung the columns 


```python
dummy_df = pd.get_dummies(train_final,drop_first=True).astype(int)
dummy_df
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
def cal_VIF(data):
    vif = pd.DataFrame()
    vif["Feature"] = data.columns
    vif["VIF"]= [variance_inflation_factor(data.values,i) for i in range(len(dummy_df.columns.to_list()))]
    return vif
    
```

```python
cal_VIF(dummy_df)
```
